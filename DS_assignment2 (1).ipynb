```python
fake_jobs.drop(fake_jobs[fake_jobs['location'].str.count(',') != 2].index, inplace=True)
fake_jobs.reset_index(inplace=True)

series = fake_jobs['location'].str.split(',', 2)
fake_jobs['city'] = 'd'
fake_jobs['state'] = 'd'
fake_jobs['country'] = 'd'
lis = series

for i in range(len(lis)):
    fake_jobs['city'][i] = lis[i][0]
    fake_jobs['state'][i] = lis[i][1]
    fake_jobs['country'][i] = lis[i][2]

fake_jobs.drop(['location'], inplace=True, axis=1)
fake_jobs.drop('index', inplace = True, axis='columns')
fake_jobs
```
